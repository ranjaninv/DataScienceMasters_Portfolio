```python
# load libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read and load stocks fundamentals csv dataset into the dataframe. 
stocks_df = pd.read_csv("fundamentals.csv")
```

```python
# Dispaly first few records
stocks_df.head(10)
```

```python
# Identify the shape of the data frame in terms of numbers of rows and columns. 
stocks_df.shape
```

```python
# Rename all column Ticker Symbol to Company_Trading_Symbol
stocks_df.columns = [c.replace(' ', '_') for c in stocks_df.columns]
stocks_df.head()
```

```python
# Identify and eliminate duplicates
print("Ticket Symbol has duplicates - {}".format(any(stocks_df.Ticker_Symbol.duplicated())))
```

```python
# Keep latest year's filing record and remove all other records for each ticker symbol 
stocks_df = stocks_df.drop_duplicates(subset=['Ticker_Symbol'], keep='last')
```

```python
stocks_df.shape
```

```python
# Drop na
stocks_df.dropna()
```

```python
# Show all the columns of the dataframe
pd.set_option('max_columns', None)
stocks_df
```

```python
# Drop rows that are not of any value for the analysis or have a large % of missing data
stocks_df = stocks_df.drop(["Unnamed:_0", "Equity_Earnings/Loss_Unconsolidated_Subsidiary", "Minority_Interest", "Misc._Stocks", "Non-Recurring_Items", "Other_Current_Liabilities", "Other_Operating_Items", "Research_and_Development"], axis = 1)
```

```python
def missing_values_table(stocks_df):
        zero_val = (stocks_df == 0.00).astype(int).sum(axis=0)
        mis_val = stocks_df.isnull().sum()
        mis_val_percent = 100 * stocks_df.isnull().sum() / len(stocks_df)
        misval_table = pd.concat([zero_val, mis_val, mis_val_percent], axis=1)
        misval_table = misval_table.rename(
        columns = {0 : 'Zero Values', 1 : 'Missing Values', 2 : '% of Total Values'})
        misval_table['Total Zero Missing Values'] = misval_table['Zero Values'] + misval_table['Missing Values']
        misval_table['% Total Zero Missing Values'] = 100 * misval_table['Total Zero Missing Values'] / len(stocks_df)
        misval_table['% Total Missing Values'] = 100 * misval_table['Missing Values'] / len(stocks_df)
        misval_table['Data Type'] = stocks_df.dtypes
        misval_table = misval_table[
            misval_table.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(stocks_df.shape[1]) + " columns and " + str(stocks_df.shape[0]) + " Rows.\n"      
            "There are " + str(misval_table.shape[0]) +
              " columns that have missing values.")
        print ("There are ", sum(misval_table['% Total Missing Values'] > 50), "columns having greater than 50% missing value.")
        print ("There are ", sum(misval_table['% Total Missing Values'] > 40), "columns having greater than 40% missing value.")
        print ("There are ", sum(misval_table['% Total Missing Values'] > 30), "columns having greater than 30% missing value.")
        print ("There are ", sum(misval_table['% Total Missing Values'] > 20), "columns having greater than 20% missing value.")
        print ("There are ", sum(misval_table['% Total Missing Values'] > 10), "columns having greater than 10% missing value.")

        return misval_table
```

```python
# Check for missing values in the stocks dataset
missing_values_table(stocks_df)
```
