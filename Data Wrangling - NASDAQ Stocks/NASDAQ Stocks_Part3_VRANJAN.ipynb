```python
# load libraries 
import pandas as pd
import numpy as np
import json
import requests
```

```python
# This function is to call API for each stock ticker symbol

def callAPI():
    
    url = "https://financialmodelingprep.com/api/v3/quotes/NASDAQ?apikey=2237da15871c0012f78a4d629e6c43a7"
    # Set header for the http request
    querystring = ("")
    headers = {'cache-control':'no-cache'}
        
    # connect and receive the response back from API
    response = requests.request("GET", url, headers=headers, params=querystring)
    if response.status_code==200:
        print("Successful Request")
        # Parse JSON response to text
        response_text = json.loads(response.text)
        print(response_text)
        return response_text
    else:
        print("Bad Response")
        return None  
```

```python
# This function is to parse the returned JSON data from the API call

def parse_json(json_data):
    
    # Define dictionary to house json data 
    nasdaq_dict = {'symbol':[], 'name':[], 'price':[], 'changesPercentage':[], 'change':[], 'dayLow':[], 'dayHigh':[], 
              'yearHigh':[], 'yearLow':[], 'marketCap':[], 'priceAvg50':[], 'priceAvg200':[], 'volume':[], 'avgVolume':[],
              'exchange':[], 'open':[], 'previousClose':[], 'eps':[], 'pe':[], 'earningsAnnouncement':[], 
              'sharesOutstanding':[], 'timestamp':[]}
    
    # Read through response JSON data and fetch each data attribute into dictiona
    for d in json_data:
        nasdaq_dict['symbol'].append(d['symbol'])
        nasdaq_dict['name'].append(d['name'])
        nasdaq_dict['price'].append(d['price'])
        nasdaq_dict['changesPercentage'].append(d['changesPercentage'])
        nasdaq_dict['change'].append(d['change'])
        nasdaq_dict['dayLow'].append(d['dayLow'])
        nasdaq_dict['dayHigh'].append(d['dayHigh'])
        nasdaq_dict['yearHigh'].append(d['yearHigh'])
        nasdaq_dict['yearLow'].append(d['yearLow'])
        nasdaq_dict['marketCap'].append(d['marketCap'])
        nasdaq_dict['priceAvg50'].append(d['priceAvg50'])
        nasdaq_dict['priceAvg200'].append(d['priceAvg200'])
        nasdaq_dict['volume'].append(d['volume'])
        nasdaq_dict['avgVolume'].append(d['avgVolume'])
        nasdaq_dict['exchange'].append(d['exchange'])
        nasdaq_dict['open'].append(d['open'])
        nasdaq_dict['eps'].append(d['pe'])
        nasdaq_dict['sharesOutstanding'].append(d['sharesOutstanding'])
        nasdaq_dict['timestamp'].append(d['timestamp'])
        
    #  Convert populated dictionary with index into dataframe
    nasdaq_df = pd.DataFrame.from_dict(nasdaq_dict, orient='index')  
    return nasdaq_df
```

```python
# Connect and fetch NASDAQ stocks Data from the API
# Load data from API into JSON format
apidata = callAPI()
```

```python
# Invoke function to parse JSON data and populate into a dataframe.
nasdaq_df = parse_json(apidata)
nasdaq_df.head()
```
